## matplotlib.pyplot

```python
### 필요한 라이브러리.모듈 임폴트
import matplotlib.pyplot as plt
```

```python
### 그래프 생성 → 환경 설정(크기, 글꼴..) → rcParams
print(f'기본 설정 값: \n{plt.rcParams}')
```

```python
### 환경 변수 → 그래프의 크기(기본값) → 재설정
## 딕셔너리의 값 재설정
plt.rcParams['figure.figsize']=[4,3]

# 결과 확인
print(plt.rcParams['figure.figsize'])
```

### 직선 그래프

```python
# 데이터 생성
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# 그래프 생성
plt.plot(x,y)

# 그래프 출력
plt.show()
```

```python
# 데이터 생성
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 8, 27, 64]

# 그래프 생성 전 → 그래프의 크기 재설정
plt.figure(figsize=(4,3))

# 그래프 생성 및 출력 → plt.show(X) → 두 그래프 중첩
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()

print('-'*80)

# 그래프 생성 전 → 그래프의 크기 재설정
plt.figure(figsize=(6,6))

# 그래프 생성 → 각각 그래프 출력
plt.plot(x,y1)
plt.show()

print('-'*80)

plt.plot(x,y2)
plt.show()
```

```python
### 직선의 색깔 지정

# 데이터 → x, y1 사용

# plot(color='색') → 그래프 생성
plt.plot(x,y1,color='#0066FF')
# plt.pot(x,y, color='g')
plt.show()
```

```python
### 직선의 모양 지정

# 데이터 생성 → x, y1

# plt.plot(x,y1,linestyle='-') → 그래프 생성
# plt.plot(x,y1,linestyle='-.')
plt.plot(x,y1,linestyle=':')
plt.show()
```

```python
### 축 레이블 설정하기

# 데이터 생성 → x, y1

# 직선 그래프 생성
plt.plot(x,y1)

# 축 레이블 설정
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## seaborn

```python
### 필요한 라이브러리 임폴트
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### 데이터 생성 --> tips 데이터


'''
1. seaborn 제공하는 데이터 : sns.load_dataset('tips')
2. tips 컬럼 설명
total_bill: 총 합계 요금표
tip: 팁
sex: 성별
smoker: 흡연자 여부
day: 요일
time: 식사 시간
size: 식사 인원
'''


# 데이터 다운로드
df_tips = sns.load_dataset('tips')


# 결과 확인하기
print(df_tips)
```

### 직선 그래프

```python
### 요금의 총합과 팁의 관계 시각화

# 그래프 생성 → lineplot()
sns.lineplot(df_tips, x='total_bill', y='tip')
plt.show()

print('-'*80)

# hue → 추가 분석 자동화, 'time' 적용
sns.lineplot(df_tips, x='total_bill', y='tip', hue='time')
plt.show()
```

```python
### 보충 : hue 기능을 사용하지 않고 식사 시간대 별 분할된 요금의 총합과 팁의 관계를 시각화 하는 경우

# 전체 데이터 → 식사 시간별대로 분할
condition1 = df_tips.loc[:,'time']=='Lunch'
condition2 = df_tips.loc[:,'time']=='Dinner'

# 전체 컬럼 지정
df_lunch = df_tips.loc[condition1,:]
df_dinner = df_tips.loc[condition2,:]

# 데이터프레임 확인하기
print(df_lunch)
print('-'*80)
print(df_dinner)

print('-'*80)

# 시각화 처리하기
sns.lineplot(data=df_lunch, x='total_bill', y='tip')
plt.show()
sns.lineplot(data=df_dinner, x='total_bill', y='tip')
plt.show()

# 매개변수 hue를 통해 한 번에 확인 가능
```

### 막대 그래프

```python
### 시각화 목표 : (범주형 데이터) 영업일 별 빈도수 + 손님의 성별

# X축에 수직인 막대 그래프 생성
sns.countplot(data=df_tips,x='day', hue='sex', palette='husl')
plt.show()

print('-'*80)

# Y축에 수직인 막대 그래프 생성
sns.countplot(data=df_tips,y='day', hue='sex', palette='Set2')
plt.show()
```

```python
### 시각화 목표 : 손님의 성별 + 영업일 별 빈도수

# X축에 수직인 막대 그래프 생성
sns.countplot(data=df_tips,x='sex', hue='day', palette='husl')
plt.show()

print('-'*80)

# Y축에 수직인 막대 그래프 생성
sns.countplot(data=df_tips,y='sex', hue='day', palette='Set2')
plt.show()
```

### 히스토그램

```python
### 시각화 목표 : tip의 금액 별 빈도수

# tip의 금액별 빈도수 → 통계 추출
tip_counts = df_tips.loc[:,'tip'].value_counts()
print(tip_counts)

print('-'*80)

# 그래프 생성 및 출력
sns.histplot(data=df_tips,x='tip',bins=100)
plt.show()

print('-'*80)

# hue = 'sex' → 성별과 팁의 관계 분석
sns.histplot(data=df_tips, x='tip', hue='sex')
plt.show()

print('-'*80)

# hue = 'time' → 식사 시간과 팁의 관계 분석
sns.histplot(data=df_tips, x='tip', hue='time')
plt.show()
```

### 박스플롯

```python
### 시각화 목표 : 요일별 팁의 관계

# 팁의 분포 시각화 (팁 그 자체에 대한 분석)
sns.boxplot(data = df_tips, y='tip')
plt.show()

print('-'*80)

# 요일별 팁의 분포 시각화(세분화)
sns.boxplot(data=df_tips,y='tip',hue='day')
plt.show()

print('-'*80)

# 요일 항목 추가 → 남는 축 → X
sns.boxplot(data=df_tips,y='tip',x='day', palette='Set2')
plt.show()

print('-'*80)

# hue = 'sex' 추가
sns.boxplot(data=df_tips,y='tip',x='day',hue='sex')
plt.show()
```

### 산점도 그래프

```python
### 시각화 목표 : 요금 총합과 팁의 관계(상관 관계) 시각화

# 그래프 생성 및 출력
sns.scatterplot(data=df_tips,x='total_bill',y='tip')
plt.show()

# 추가 분석 : 상관 계수 측정 → df.corr()
print(df_tips.loc[:,['total_bill','tip']].corr())

print('-'*80)

# hue = 'time' → 추가 분석
sns.scatterplot(data=df_tips,x='total_bill',y='tip',hue='time')
plt.show()

print('-'*80)

# 추가 분석 : 전체 데이터 → 식사 시간대별로 분할
condition1 = df_tips.loc[:,'time']=='Lunch'
condition2 = df_tips.loc[:,'time']=='Dinner'

df_lunch = df_tips.loc[condition1,:]
df_dinner = df_tips.loc[condition2,:]

lunch_corr = df_lunch.loc[:, ['total_bill','tip']].corr()
dinner_corr = df_dinner.loc[:, ['total_bill', 'tip']].corr()

print(f'점심 시간대의 요금 총합과 팁의 상관 계수 : \n{lunch_corr}')
print('-'*80)
print(f'저녁 시간대의 요금 총합과 팁의 상관 계수 : \n{dinner_corr}')
```

### 히트맵

```python
### 상관 분석 실행 → 상관 행렬 생성

# 상관 분석 → df.corr()
corr_matrix = df_tips.corr(numeric_only=True)
print(f'상관 분석의 결과 : \n{corr_matrix}')

print('-'*80)

# heatmap 그래프 생성 및 출력
sns.heatmap(data=corr_matrix, annot=True, fmt='.2f')
plt.show()
```

## pandas

```python
### 연도별 직원 수 시각화

# 필요한 라이브러리 임폴트
import pandas as pd

# 데이터 생성
year = ['2017','2018','2019']
number = [10, 40, 90]

# pd.Series() 사용 → Series 자료형 데이터 생성
s = pd.Series(data = number, index = year)
print(f'생성된 Series 데이터 확인 \n{s}')

print('-'*80)

# 직선 그래프 생성 및 출력
s.plot(kind='line')
# plt.show() : plt.show()를 두 개의 그래프 마지막에 작성한다는 것은 중첩된 그래프를 작성하겠다는 의미

# 막대 그래프 생성 및 출력
s.plot(kind='bar')
plt.show()

# 막대 그래프 생성 및 출력
s.plot(kind='barh')
plt.show()
```
